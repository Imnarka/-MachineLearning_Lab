Как влияет пол на выживание?

```python
df_survived = df[df.Survived==1]
print(df_survived.groupby('Sex')['Survived'].count())
```

Влияние класса на выживаемость

```python
survived_by_class = df.groupby('Pclass').agg(['count', 'sum'])['Survived'] # считаем кол-во всех и сумму уже выживших
survived_by_class_percent = survived_by_class['sum'] / survived_by_class['count'] * 100
result = pd.concat([survived_by_class, survived_by_class_percent], axis=1)
result.columns = ['Count','Survived','Survive %']
result
```
